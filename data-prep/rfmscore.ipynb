```python
import pandas as pd
import numpy as np
import math
```

```python
rfm_data = pd.read_csv('data/RFM_clean.csv')
rfm_data.head()
```

```python
rfm_data["Monetary"] = rfm_data["Monetary"].apply(lambda x : int(math.floor(x)))
rfm_data.head()
```

```python
print(rfm_data[rfm_data['Monetary'].isnull()])
```

```python
rfm_data["R"] = pd.cut(rfm_data["Recency"], 
                                       bins=[-1, 
                                             np.percentile(rfm_data["Recency"], 20), 
                                             np.percentile(rfm_data["Recency"], 40), 
                                             np.percentile(rfm_data["Recency"], 60),
                                             np.percentile(rfm_data["Recency"], 80),
                                             rfm_data["Recency"].max()], 
                                       labels=[5, 4, 3, 2, 1]).astype("int")

rfm_data["F"] = pd.cut(rfm_data["Frequency"], 
                                       bins=[-1, 
                                             np.percentile(rfm_data["Frequency"], 20), 
                                             np.percentile(rfm_data["Frequency"], 40), 
                                             np.percentile(rfm_data["Frequency"], 60),
                                             np.percentile(rfm_data["Frequency"], 80),
                                             rfm_data["Frequency"].max()], 
                                       labels=[1, 2, 3, 4, 5]).astype("int")

rfm_data["M"] = pd.cut(rfm_data["Monetary"], 
                                       bins=[-1, 
                                             np.percentile(rfm_data["Monetary"], 20), 
                                             np.percentile(rfm_data["Monetary"], 40), 
                                             np.percentile(rfm_data["Monetary"], 60),
                                             np.percentile(rfm_data["Monetary"], 80),
                                             rfm_data["Monetary"].max()], 
                                       labels=[1, 2, 3, 4, 5]).astype("int")                                       


rfm_data.head()
```

```python
rfm_score = rfm_data[['R','F','M']]
rfm_score.head()
```

```python
rfm_score.to_csv('data/RFM_score.csv', index=False)
```
